# Bump-on-Tail

```python
# set up configuration parameters
# number of mesh points in x
Nx = 20
# number of spectral expansions
Nv = 60
# Velocity scaling of electron and ion
alpha_e1 = np.sqrt(2)
alpha_e2 = np.sqrt(1 / 2)
alpha_i = np.sqrt(2 / 1863)
# perturbation magnitude
epsilon = 0.03
# x grid is from 0 to L
L = 20 * np.pi / 3
# final time
T = 10.
# time stepping
dt = 1e-3
# time vector
t_vec = np.linspace(0, T, int(T / dt) + 1)
# velocity scaling
u_e1 = -0.5
u_e2 = 4.5
u_i = 0
# mass normalized
m_e1 = 1
m_e2 = 1
m_i = 1863
# charge normalized
q_e1 = -1
q_e2 = -1
q_i = 1
# scaling of bulk and bump
delta_e1 = 9/10
delta_e2 = 1/10
```

```python
# velocity grid for plotting purposes
v = np.linspace(-5, 8, int(1e5))
# spatial grid 
x = np.linspace(0, L, 2*Nx+1)
```

```python
# initial condition of the first expansion coefficient
C_0i = np.zeros(2 * Nx + 1)
C_0i[Nx] = np.sqrt(1 / alpha_i)
```

```python
# initial condition of the first expansion coefficient
C_0e1 = np.zeros(2 * Nx + 1, dtype="complex128")
C_0e2 = np.zeros(2 * Nx + 1, dtype="complex128")

# project the electron species onto fourier space
for ii, kk in enumerate(range(-Nx, Nx + 1)):
    C_0e1[ii] = np.trapz(y=np.sqrt(delta_e1 * (1 + epsilon * np.cos(0.3 * x)) / alpha_e1) * np.exp(-2 * np.pi * 1j * kk * x / L),
                         x=x,
                         dx=x[1] - x[0]) / L
    C_0e2[ii] = np.trapz(y=np.sqrt(delta_e2 * (1 + epsilon * np.cos(0.3 * x)) / alpha_e2) * np.exp(-2 * np.pi * 1j * kk * x / L),
                         x=x,
                         dx=x[1] - x[0]) / L
```

```python
# initialize states (electrons and ions)
states_e1 = np.zeros((Nv, Nx * 2 + 1), dtype="complex128")
states_e2 = np.zeros((Nv, Nx * 2 + 1), dtype="complex128")
states_i = np.zeros((Nv, Nx * 2 + 1), dtype="complex128")

# initialize the expansion coefficients
states_e1[0, :] = C_0e1
states_e2[0, :] = C_0e2
states_i[0, :] = C_0i
```

```python
E_coeff = solve_poisson_equation_two_stream(state_e1=states_e1,
                                            state_e2=states_e2,
                                            state_i=states_i,
                                            alpha_e1=alpha_e1,
                                            alpha_e2=alpha_e2,
                                            alpha_i=alpha_i,
                                            Nx=Nx,
                                            L=L)
```

```python
# fft of electric field 
E_field= fft_(coeff=E_coeff, Nx=Nx, x=x, L=L)
```

```python
plt.plot(x, E_field, linewidth=5, color="yellow", label="E")
plt.plot(x, -1e-1*np.sin(x*0.3), color="black", label="Correct")
```

```python
# the distribution function of electrons and protons for plotting purposes
f_0e1 = np.outer(fft_(C_0e1, Nx, x, L), psi_ln_sw((v-u_e1)/alpha_e1, n=0)).real
f_0e2 = np.outer(fft_(C_0e2, Nx, x, L), psi_ln_sw((v-u_e2)/alpha_e2, n=0)).real
f_0i = np.outer(fft_(C_0i, Nx, x, L), psi_ln_sw((v-u_i)/alpha_i, n=0)).real
```

```python
sol_midpoint_u = np.load("data/SW_sqrt/bump_on_tail/ampere/sol_midpoint_u_60.npy")
sol_midpoint_t = np.load("data/SW_sqrt/bump_on_tail/ampere/sol_midpoint_t_60.npy")
print(np.shape(sol_midpoint_u))
print(np.shape(sol_midpoint_t))
```

```python
# initialize the electric potential
E_midpoint = np.zeros((2*Nx+1, len(sol_midpoint_t)), dtype="complex128")
```

```python
# initialize the states for implicit midpoint (symplectic)
state_e1_midpoint = np.zeros((Nv, 2*Nx+1, len(sol_midpoint_t)), dtype="complex128")
state_e2_midpoint = np.zeros((Nv, 2*Nx+1, len(sol_midpoint_t)), dtype="complex128")
state_i_midpoint = np.zeros((Nv, 2*Nx+1, len(sol_midpoint_t)), dtype="complex128")


for ii in range(len(sol_midpoint_t)):
    for jj in range(Nv):
        # unwind the flattening in order to solve the linear system
        state_e1_midpoint[jj, :, ii] = sol_midpoint_u[jj * (2 * Nx + 1): (jj + 1) * (2 * Nx + 1), ii]
        state_e2_midpoint[jj, :, ii] = sol_midpoint_u[Nv * (2 * Nx + 1) + jj * (2 * Nx + 1): Nv * (2 * Nx + 1) + (jj + 1) * (2 * Nx + 1), ii]
        state_i_midpoint[0, Nx, ii] = np.sqrt(1 / alpha_i)
        
    E_midpoint[:Nx+1, ii] = sol_midpoint_u[2 * Nv * (2 * Nx + 1): (2*Nv +1) * (2 * Nx + 1), ii][:Nx+1]
    E_midpoint[Nx+1:, ii] = np.flip(np.conjugate(E_midpoint[:Nx, ii]))
```

```python
E_field = np.zeros(np.shape(E_midpoint))
for ii in range(len(sol_midpoint_t)):
    E_field[:, ii] = fft_(E_midpoint[:, ii], Nx, x, L)
```

# Conservation Laws

### mass conservation 

$$
N^{s}(t) = \alpha^{s} \sum_{n=0}^{N_{v}-1} \int_{0}^{L} C_{n}^{s}(x, t)^{2}\mathrm{d}x =  \alpha^{s} \sum_{n=0}^{N_{v}-1} \int_{0}^{L} \sum_{k=-N_{x}}^{N_{x}} \sum_{m=-N_{x}}^{N_{x}} C_{n, k}^{s}(t) C_{n, m}^{s}(t) \exp(2\pi x i k/ L) \exp(2\pi x i m/ L) \mathrm{d}x
= \alpha^{s} L \sum_{n=0}^{N_{v}-1} \sum_{k=-N_{x}}^{N_{x}} C_{n, k}^{s}(t) C_{n, -k}^{s}(t)
$$


```python
def integral(k, m):
    return -1j*L*(1+np.exp(2*np.pi*1j*(k+m)))/(2*(k+m)*np.pi)
```

```python
def mass(state):
    y = 0
    for nn in range(0, Nv):
        y += L * np.convolve(state[nn, :], state[nn, :], mode="same")
    return y 
```

```python
mass_midpoint = np.zeros(len(sol_midpoint_t))

for ii in range(0, len(sol_midpoint_t)):
    mass_midpoint[ii] = mass(state_e1_midpoint[:, :, ii])[Nx].real*alpha_e1\
                      + mass(state_e2_midpoint[:, :, ii])[Nx].real*alpha_e2\
                      + mass(state_i_midpoint[:, :, ii])[Nx].real*alpha_i
```

### momentum conservation

```python
def ampere_term(state):
    """
    :param state: coefficient of species s
    :return: quadratic term
    """
    Nv, Nx = np.shape(state)
    res = np.zeros(Nx, dtype="complex128")
    for m in range(1, Nv):
        res += np.sqrt(m) * np.convolve(a=state[m, :], v=state[m-1, :], mode="same")
    return res

def density_convolve(state):
    """ state of all coefficients, dimensions = Nv x (2*Nx+1)

    :param state:
    :return:
    """
    Nv, Nx_total = np.shape(state)
    y = np.zeros(Nx_total, dtype="complex128")
    for ii in range(Nv):
        y += np.convolve(a=state[ii, :], v=state[ii, :], mode="same")
    return y
```

```python
momentum_midpoint = np.zeros(len(sol_midpoint_t), dtype="complex128")

for ii in range(0, len(sol_midpoint_t)):
    # first term
    term1 = alpha_e1 * u_e1 * density_convolve(state=state_e1_midpoint[:, :, ii])[Nx]
    term2 = alpha_e2 * u_e2 * density_convolve(state=state_e2_midpoint[:, :, ii])[Nx]
    term3 = alpha_i * u_i * density_convolve(state=state_i_midpoint[:, :, ii])[Nx]
    # second term
    term4 = np.sqrt(2) * (alpha_e1 ** 2) * ampere_term(state=state_e1_midpoint[:, :, ii])[Nx]
    term5 = np.sqrt(2) * (alpha_e2 ** 2) * ampere_term(state=state_e2_midpoint[:, :, ii])[Nx]
    term6 = np.sqrt(2) * (alpha_i ** 2) * ampere_term(state=state_i_midpoint[:, :, ii])[Nx]

    momentum_midpoint[ii] = L * (q_e1 * (term1 + term4) + q_e2 * (term2 + term5) + q_i * (term3 + term6))
```
